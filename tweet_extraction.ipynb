```python
import tweepy
import pandas
import json
import re
import itertools
```

```python
tweet_ids = []
with open('paristweetids.txt','r') as f:
    tweet_ids.append(f.read().split('\n'))
```

```python
tweet_ids = list(itertools.chain.from_iterable(tweet_ids))
```

```python
len(tweet_ids)
```

```python
auth = tweepy.OAuthHandler('Consumer API Key', 'Consumer API secret key')
auth.set_access_token('access token key', 'access token secret key')
api = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)

```

```python
tweet_id = []
tweet = []
profile_id = []
profile_name = []
screen_name = []
friends_count = []
created_at = []
listed_count = []
favorites_count = []
statuses_count = []
user_time_zone = []
followers_count = []
user_lang = []
verified = []
followers_count = []
location = []
retweets = []
description = []
```

```python
for j in range(0,,len(tweet_ids),100):
    
    tweets = api.statuses_lookup(tweet_ids[j:j+100]) # id_list is the list of tweet ids
    try:
    #tweet_txt = []
        for i in tweets:
            tweet_id.append(i.id)
            tweet.append(i.text)
            profile_id.append(i.user.id)
            profile_name.append(i.user.name)
            friends_count.append(i.user.friends_count)
            created_at.append(i.user.created_at)
            listed_count.append(i.user.listed_count)
            favorites_count.append(i.user.favourites_count)
            statuses_count.append(i.user.statuses_count)
            verified.append(i.user.verified)
            screen_name.append(i.user.screen_name)
            user_time_zone.append(i.user.time_zone)
            followers_count.append(i.user.followers_count)
            user_lang.append(i.user.lang)
            description.append(i.user.description)
            retweets.append(i.retweet_count)
            location.append(i.user.location)
            followers_count.append(i.user.followers_count)
        print(j)
    except Exception:
        #print(j)
        pass
    
        
```
